```python
math_keywords = [
    "mathematics",
    "calculus",
    "algebra",
    "statistics",
    "probability",
    "analysis",
    "topology",
    "graph theory",
    "optimization",
    "machine learning"
]

books["genres"] = books["genres"].fillna("").str.lower()

math_books = books[
    books["genres"].apply(
        lambda x: any(k in x for k in math_keywords)
    )
]
```
