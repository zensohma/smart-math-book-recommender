```python
import pandas as pd

books = pd.read_csv("../data/raw/books.csv")

print("Before:", books.shape)

books = books[
    [
        "bookId",
        "title",
        "author",
        "genres",
        "description"
    ]
]

books = books.drop_duplicates()

books = books.dropna(
    subset=[
        "title",
        "author",
        "description"
    ]
)

print("After:", books.shape)

books.head()
```

```python
books["content"] = (
    books["title"].astype(str)
    + " "
    + books["author"].astype(str)
    + " "
    + books["description"].astype(str)
)
```

```python
books["content"].iloc[0]
```

```python
books.to_csv(
    "../data/processed/books_for_ml.csv",
    index=False
)
```
